# 4 그룹별 데이터 분포 시각화
seaborn의 plot 종류 사용하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/CV_work/AICE공부/data_csv/food_saltiness.csv')
df.info()

df.head(5)
```

```python
# boxplot -> 정보 요약

# 도화지 생성

plt.figure(figsize=(10,6))

# boxplot사용
sns.boxplot(x='main_ingredient', y='saltiness',data=df,hue = 'cooking_method' )
# hue는 해당 변수에 따라 색상으로 구분

# 그래프 제목 붙이고 시각화
plt.title('name')
plt.show()

# 상자는 중간수중 50%가 분포하는 핵심 범위
# 상자 내부 선이 median값
# 위아래 선은 상한선과 하한선
# 이상치는 그 밖에 있을수도 있고 점으로 표시됨
```

```python
# barplot -> 평균만 알고싶을 때
#  boxplot은 튀는값을 이상치로 처리하는데, barplot은 이상치 처리를 따로 안함.

sns.barplot(x='main_ingredient', y='saltiness',data=df,errorbar=('ci',95))
# errorbar 는 평균값의 불확실성이라는데 모르겠음.


plt.title('barplot')

plt.show()
```

```python
# violinplot 전체 스펙트럼

sns.violinplot(x='main_ingredient', y='saltiness',data=df,hue= 'cooking_method',split = True)
# split = True하면 같은 x축을 사용함.

plt.title('violin')
plt.show()
```

```python
# countplot -> 갯수만 확인
sns.countplot(x='main_ingredient',data=df)
plt.title('count')
plt.show()
```

```python
# catplot 얘는 box, violin, bar 다 한번에 할 수 있음.

g = sns.catplot(x='main_ingredient', y='saltiness',data=df,
                col = 'cooking_method', # col로 데이터 분할
                kind='box', # boxplot사용
                height = 5 , aspect=1 # subplot크기
                )

# 전체 제목 설정
g.fig.suptitle('Saltiness bt Ingredient(Split by Cooking Method)')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x='main_ingredient', y='saltiness',data=df)
sns.stripplot(x='main_ingredient', y='saltiness',data=df, color = 'black',alpha = 0.5)
# stripplot은 데이터 개수를 점으로 찍어줌. 먼소린지는 모르겠음
# alpha는 투명도.
plt.title('dot')
plt.show()
```

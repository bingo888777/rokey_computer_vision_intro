# 경사하강법과 옵티마이저 구현하기

###0.시작하기 전 이론 살펴보기 (복습)

####0-1. 경사하강법이란?

경사 하강법은 함수의 기울기(경사)를 사용하여 함수의 최솟값을 찾는 방법입니다. 즉, 비용 함수(Cost Function)의 기울기를 계산하여 기울기의 반대 방향으로 조금씩 이동시키면서 최적의 파라미터를 찾아가는 과정입니다.

####0-2. 경사하강법 기본 원리

1. 파라미터 초기화: 모델의 파라미터를 임의의 값으로 초기화합니다.
2. 비용 함수 계산: 현재 파라미터에서의 비용(또는 손실)을 계산합니다.
3. 기울기 계산: 비용 함수의 기울기(파라미터에 대한 미분값)를 계산합니다.
4. 파라미터 업데이트: 계산된 기울기 방향의 반대로 파라미터를 조금씩 조정합니다.
5. 수렴 확인: 비용이 충분히 낮아지거나, 지정된 반복 횟수에 도달할 때까지 2~4단계를 반복합니다.

###1.파이썬으로 경사하강법 구현하기



```python
import numpy as np

# 예제 데이터
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# 파라미터 초기화
W = 0.0  # 가중치
b = 0.0  # 편향

# 학습률과 반복 횟수 설정
learning_rate = 0.01
iterations = 1000

# 비용 함수
def cost_function(X, y, W, b):
    """MSE = Mean Squared Error"""
    n = len(X)
    cost = np.sum((W*X + b - y)**2) / n
    return cost

# 비용 함수의 기울기 계산 함수
def gradient(X, y, W, b):
    n = len(X)
    dW = np.sum(2*X*(W*X + b - y)) / n
    db = np.sum(2*(W*X + b - y)) / n
    return dW, db

# 경사 하강법
for i in range(iterations):
    dW, db = gradient(X, y, W, b)
    W = W - learning_rate * dW
    b = b - learning_rate * db

    # 100회 반복마다 비용 출력
    if i % 100 == 0:
        cost = cost_function(X, y, W, b)
        print(f"Iteration {i}: Cost = {cost}, W = {W}, b = {b}")

# 최종 결과 출력
print(f"Final parameters: W = {W}, b = {b}")
```

- `cost_function`: 비용 함수를 계산합니다. 여기서 비용은 실제 값과 예측 값의 차이의 제곱합의 평균입니다.
- `gradient`: 비용 함수의 기울기를 계산합니다. 이 기울기는 가중치(W)와 편향(b)에 대한 비용 함수의 미분 값입니다.
- `경사 하강법 구현`: 반복 횟수만큼 기울기를 계산하고, 이를 사용하여 파라미터(W와 b)를 업데이트합니다. 파라미터는 학습률(learning_rate)에 비례하여 기울기의 반대 방향으로 조정됩니다. 이 과정을 통해 비용 함수의 최솟값을 찾아가며, 최적의 파라미터를 결정합니다.

### 2.경사하강법 더 알아보기

추가적으로 아래 2가지를 참고하여 직접 실습해볼 수 있습니다.

1. 학습률(Learning Rate)의 중요성

2. 수렴 확인
<br> 경사 하강법은 일정 조건을 만족하면 `종료`됩니다. 이 조건은 일반적으로 `비용 함수의 값이 더 이상 크게 변하지 않거나, 지정된 최대 반복 횟수에 도달했을 때`입니다. 코드 예제에서는 최대 반복 횟수를 기준으로 사용했습니다.



###3.SGD (Stochastic Gradient Descent)

SGD는 가장 기본적인 최적화 알고리즘으로, 각 스텝에서 무작위로 선택된 하나의 데이터 샘플에 대한 기울기를 계산하여 모델을 업데이트합니다.

```python
class SGD:
    def __init__(self, learning_rate=0.01):
        self.lr = learning_rate

    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key]
```

- 이 코드는 파라미터를 업데이트하는 부분입니다. 학습률(self.lr)과 그래디언트(grads[key])의 곱을 현재 파라미터 값에서 빼주어 파라미터를 업데이트합니다. 이는 경사 하강법의 기본 원리에 따라, 비용 함수의 기울기 반대 방향으로 조금씩 이동시키는 과정을 구현한 것입니다.

###4.Momentum


Momentum은 이전 업데이트의 방향을 기억하여 그 방향으로 일정 부분 이동시키는 방식을 추가한 알고리즘입니다. 이는 공이 그릇을 구르듯이 최적점을 향해 가속도를 붙여줍니다.



```python
class Momentum:
    def __init__(self, learning_rate=0.01, momentum=0.9):
        self.lr = learning_rate
        self.momentum = momentum
        self.v = None

    def update(self, params, grads):
        if self.v is None:
            self.v = {}
            for key, val in params.items(): # params.items() params 는 dict라서 key,value가 반환.
                self.v[key] = np.zeros_like(val) # params를 복사하는데, value는 전부 0으로 초기

        for key in params.keys():
            self.v[key] = self.momentum * self.v[key] - self.lr * grads[key]
            params[key] += self.v[key]
```

- 첫 번째 줄은 모멘텀 업데이트를 계산하는 부분입니다. 이전 업데이트의 속도(self.v[key])에 모멘텀 값(self.momentum)을 곱하고, 현재 그래디언트에 학습률을 곱한 값을 빼줍니다. 이는 공이 경사면을 내려오면서 가속도가 붙는 것과 유사한 효과를 모델 학습에 적용한 것입니다.
- 두 번째 줄에서는 계산된 속도(self.v[key])를 파라미터에 적용하여 업데이트합니다. 이 과정을 통해 이전 업데이트의 방향성을 유지하면서 파라미터를 더욱 효율적으로 최적화할 수 있습니다.

###5.RMSprop

RMSprop은 학습률을 각 파라미터에 대해 독립적으로 조정하여, 수평 방향으로는 빠르게, 수직 방향으로는 느리게 조정합니다. 이는 비등방성(방향에 따라 다른 성질을 가지는) 함수에서 유리합니다.

```python
class RMSprop:
    def __init__(self, learning_rate=0.01, decay_rate=0.99):
        self.lr = learning_rate
        self.decay_rate = decay_rate
        self.h = None

    def update(self, params, grads):
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)

        for key in params.keys():
            self.h[key] *= self.decay_rate
            # 이전 누적된 값에서 감쇠율을 곱한다 -> 기존 기울기 값 영향을 줄인다.
            self.h[key] += (1 - self.decay_rate) * (grads[key] ** 2)
            # 최근 기울기에 더 높은 가중치 부여 # 현재가 아닌 최근!
            params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7)
            # 기울기가 누적된 기울기 크기의 제곱근 으로 나누어줌(학습률을 파라미터 별로 자동조)
```

- 첫 번째 줄은 이동 평균의 가중치 감소를 적용하는 부분입니다. self.h[key]는 그래디언트의 제곱에 대한 이동 평균을 저장하며, self.decay_rate를 통해 이전 값에 대한 가중치를 조절합니다.
- 두 번째 줄에서는 현재 그래디언트의 제곱을 이동 평균에 추가합니다. 이는 그래디언트의 변화량을 안정적으로 만들어주어, 학습 과정에서 발생할 수 있는 급격한 변화를 완화시킵니다.
- 세 번째 줄에서 파라미터를 업데이트하는 과정입니다. 여기서는 self.h[key]의 제곱근을 사용하여 각 파라미터에 대한 학습률을 조정합니다. 이로 인해 각각의 파라미터가 서로 다른 속도로 업데이트될 수 있게 하여, 더 효율적인 학습이 가능하게 됩니다.

- ht : 기울기 제곱의 지수 이동 평균
- p : 감쇠율 (decay_rate) 일반적으로 0.99
- gt : 현재 기울기(gradient)
- epsilon: 아주 미세한 값 >> 여기서는 zero division 방지 (1e-7 >> 0.0000001)
- 지수이동평균(EMA) 감쇠율에 의해서 오래된 미분값(기울기)는 점점 영향력이 줄어듦

- 제곱을 했기 때문에 음수 제거 가능, 높은 기울기에 대해서 (가파름) 더 낮은 lr을 적용하고, 완만한( 낮은 기울기) 에 대해서 더 높은 lr을 적용한다.
- 제곱을 사용해서 가파를수록 훨씬 적은 lr을 갖게 된다.

###6.Adam

Adam은 Momentum과 RMSprop의 아이디어를 결합한 알고리즘으로, 모멘텀을 사용하여 과거 기울기의 지수 이동 평균을 계산하고, RMSprop처럼 기울기의 제곱에 대한 지수 이동 평균을 사용합니다.

```python
class Adam:
    def __init__(self, learning_rate=0.001, beta1=0.9, beta2=0.999):
        self.lr = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.iter = 0
        self.m = None
        self.v = None

    def update(self, params, grads):
        if self.m is None or self.v is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)

        self.iter += 1
        for key in params.keys():
            self.m[key] = self.beta1 * self.m[key] + (1 - self.beta1) * grads[key]
            self.v[key] = self.beta2 * self.v[key] + (1 - self.beta2) * (grads[key] ** 2)
            m_hat = self.m[key] / (1 - self.beta1 ** self.iter)
            v_hat = self.v[key] / (1 - self.beta2 ** self.iter)
            params[key] -= self.lr * m_hat / (np.sqrt(v_hat) + 1e-7)
```

- 첫 번째와 두 번째 줄에서는 각각 모멘텀(self.m[key])과 속도(self.v[key])를 업데이트합니다. 이는 과거 그래디언트의 지수 가중 이동 평균을 계산하는 과정입니다. self.beta1과 self.beta2는 이동 평균을 계산할 때 사용되는 감소율입니다.
- 세 번째와 네 번째 줄에서는 편향 보정이 적용된 모멘텀(m_hat)과 속도(v_hat)를 계산합니다. 이는 학습 초기에 self.m[key]와 self.v[key]가 0으로 초기화되어 있는 상황에서의 편향을 보정하기 위함입니다.
- 마지막 줄에서는 보정된 모멘텀과 속도를 사용하여 파라미터를 업데이트합니다. 여기서도 self.v[key]의 제곱근을 사용하여 각 파라미터에 대한 학습률을 동적으로 조정합니다.

- mt 1차 모멘트 : 기울기의 지수이동평균
- vt 2차 모멘트 : 기울기 제곱의 지수이동 평균
여기서의 모멘트는 -> 1차는 기울기의 방향, 2차는 기울기의 크기(이동?)
- m_t(m_hat), v_t(v_hat) : 편향 보정된 1차, 2차 모멘트
- beta1_t : beta1을 t번 곱한것. t가 커지면 0으로 수렴

정리하면,
1. 최근에 어느 방향으로 이동했는가? -> 1차 모멘트, Momentum
2. 얼마나 크게 움직였는가? -> 2차 모멘트, RPSprop
Momentum과 RPSprop 두 개를 합친것.
- 각 파라미터마다 크기와 방향을 업데이트

###7.AdamW

AdamW는 Adam 최적화 방법에 가중치 감소(weight decay)를 적용한 변형입니다. 이는 과적합을 방지하고 일반화 성능을 향상시키는 데 도움을 줍니다.



```python
class AdamW:
    def __init__(self, learning_rate=0.001, beta1=0.9, beta2=0.999, weight_decay=0.01):
        self.lr = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.weight_decay = weight_decay
        self.iter = 0
        self.m = None
        self.v = None

    def update(self, params, grads):
        if self.m is None or self.v is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)

        self.iter += 1
        for key in params.keys():
            # 편향된 첫번째 모멘트 추정치 업데이트
            self.m[key] = self.beta1 * self.m[key] + (1 - self.beta1) * grads[key]
            # 편향된 두번째 모멘트 추정치 업데이트
            self.v[key] = self.beta2 * self.v[key] + (1 - self.beta2) * (grads[key] ** 2)
            # 편향 보정된 첫번째 모멘트 추정치 계산
            m_hat = self.m[key] / (1 - self.beta1**self.iter)
            # 편향 보정된 두번째 모멘트 추정치 계산
            v_hat = self.v[key] / (1 - self.beta2**self.iter)

            # weight decay 적용
            params[key] -= self.weight_decay * params[key]

            # 파라미터 업데이트
            params[key] -= lr_t * m_hat / (np.sqrt(v_hat) + 1e-7)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

video_path = '/content/drive/MyDrive/CV_work/AI_응용_강의자료/data/greenball.mp4'
```

```python
def track(image):
    # BGR(numpy array) 입력 >> 초록색 물체 중심 좌표 (centroid) - 색상 tracking



    # 1) 노이즈 제거 > 가우시안 블러
    blur = cv2.GaussianBlur(image, (5, 5),0) # img, ksize , sigmaX : 0 -> 자동계산

    # 2) BGR >> HSV 색 공간 변환
    hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)
    # 아마 video에서 프레임마다 짤라서 캡처 이미지를 가져오겠죠?
    # 그럼 거기서 가져온 image를쓸거고, 그건 BGR이에요
    # HSV로 바꾸는 이유는 조명이 바뀌어도 색상 찾기가 쉽기 떄문에

    # 3) 초록색 범위 설정

    lower_green = np.array([25,60,100]) #저번에 blue 영역 설정할때는 채도 와 명도가 50 ~ 255였는데
                                        #여기서는 그냥 임의로 쓰는 느낌?
    upper_green = np.array([50,255,255])

    # 4) 마스크(mask) (관심영역 : ROI)
    mask = cv2.inRange(hsv, lower_green, upper_green)
    # 초록색 부분 -> True -> 255 로 만들기, 나머지는 False -> 0으로 만들기


    # 5) 마스크도 블러처리
    bmask = cv2.GaussianBlur(mask, (5,5), sigmaX = 0)

    # 6) 모멘트(moments) 이용해서 중심좌표 계싼

    moments = cv2.moments(bmask)
    # key-value {} 저장  (m00: ,..면적)
    m00 = moments['m00']
    # m00 : 전체 면적(하얀 픽셀 개수)


    centroid_x, centroid_y = None, None
    # 중심좌표 초기화 = cv2.moments(bmask)
        # key-value {} 저장  (m00: ,..면적)
    m00 = moments['m00']
    # m00 : 전체 면적(하얀 픽셀 개수)


    centroid_x, centroid_y = None, None
    # 중심좌표 초기화


    # 만약 면적이 0이 아니라면 (면적이 존재한다면)
    if m00 !=0:
       centroid_x = int(moments['m10']/m00) # 한 영역의 중심 x 좌표
       centroid_y = int(moments['m01']/m00) # 한 영역의 중심 y 좌표

       # m10, m01 의미 (1차 모멘트)
       # 1차 모멘트 : 이미지 픽셀 같이 이산 데이터(1, 0 등)에서 각 요소의 값(강도) 위치 곱한 것들의 합
       # 이미지 처리에서 1차 모멘트 계산시 모든 픽셀의 위치(x,y)는 (0,0) (맨 왼쪽 상단) 지점부터 거리
       # m10 : x축 방향에 대한 pixel 강도의 합
       # m01 : y축 방향에 대한 pixel 강도의 합
       # 즉, 해당 좌표의 픽셀의 밝기 의미
       # 참고) 0차 모멘트(m00) 어떤 존재(객체) 있냐, 그것의 크기
       # >> 객체의 질량(면적) >> 모든 픽셀 값을 다 더함

       # 중심좌표(centroid) 공식
       # cx(객체의 가로 중심 좌표) = m10 / m00(전체면적)
       # cy(객체의 세로 중심 좌표) = m01 / m00(전체면적)



    # 기본값: 중심 없음
    # >> 초록색을 못 찾으면
    # 왜 -1 을 썼나요? 유효하지 않은 좌표(x좌표, y좌표 >=0)
    # >> 즉, 이미지 화면에 보이는 것 (0,0)~(w,h)
    # center point 없어 >> 객체 못 찾았어
    # >> 객체가 있다면 (중심점 있다면) >> (cx, cy)

    # -1은 이미지 평면에 존재 안하는 숫자([0,255])
    # 좌상단 (0,0)
    crt = (-1, -1)

    # 중심이 계산된 경우
    if centroid_x != None and centroid_y != None:
      crt = (centroid_x, centroid_y)
      # 이미지 위에 중심점 표시(검은색 점)
      cv2.circle(image, crt, 4, (0,0,0), -1)
      # 4: radius(반지름), -1 : 원의 내부를 지정된 색으로 채워라

    cv2_imshow(image) #colab에서 cv2.imshow() 대신 사용

    return crt

# 메인 실행 부분 (동영상 캡처)

if __name__ == '__main__':
    # 프로그램 시작점(이 파일을 직접 실행할 때만 동작)
    capture = cv2.VideoCapture(video_path)

    if not capture.isOpened():
        print('동영상을 열 수 없습니다.')
    else:
        frame_idx = 0

        while True:
            okay, image = capture.read()

            # 비디오가 마지막이면
            if not okay:
                print('영상 끝까지 재생 완료')
                break

            # 초록색 공 추적
            ctr = track(image)
            print(f'Frame {frame_idx}: centroid = {ctr}')

            frame_idx += 1

        capture.release()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # 콜랩에서 이미지 출력 지원

video_path = '/content/drive/MyDrive/두산로보틱스_딥러닝_컴퓨터비전/8기/컴퓨터비전_AI응용/data/greenball.mp4'
```

```python
def track(image):
    # BGR 이미지(numpy array) 입력  >> 초록색 물체 중심좌표(centroid)추적

    # 1)노이즈 제거 >> 블러(blur)처리(가우시안 블러)
    blur = cv2.GaussianBlur(image, (5,5),0)
    # (5,5) 5*5 filter, 0 : 표준편차 자동계산

    # 2) BGR >> HSV 색 공간 변환
    hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)
    # HSV(Hue, Saturation, Value)
    # hue: 색상(빨강, 초록, 파랑...) / saturation: 채도(색이 진하니?) / value(명도, 빛의 밝기)
    # HSV 로 왜 바꾸나요? 조명이 바뀌어도 색상 찾기가 더 쉬움
    # BGR: 컴퓨터 중심(색 검출이 힘들어요. 파랑, 그린, 빨강 위주임. 조명의 밝기 영향을 많이 받아요)

    # 3) 초록색 범위 설정(필요하면 값 조정 가능)
    # 연한 녹색 - 진한 녹색 범위 설정
    lower_green = np.array([40, 70, 70]) # 색상 40, 채도 70, 명도 70
    upper_green = np.array([80, 200, 200])

    # 4) 마스크(mask) 생성
    mask = cv2.inRange(hsv, lower_green, upper_green)
    # 초록색 부분만 하얀색 (255), 나머지는 검은색(0) 만들기
    # 초록색 부분만 오려내는 효과 발생

    # lower green - pixel value - upper green
    # inRange 이 범위 안에 있으면 >> 255(흰색) / 없으면 0 (검은색)

    # 5) 마스크를 다시 블러링(잡음 noise 제거)
    bmask = cv2.GaussianBlur(mask, (5,5),0)
    # bmask (binary mask) 이미지가 0(code 0) 또는 255(code 1) 구성된 영상 (흑백영상)

    # 6) 모멘트(moments) 이용하여 중심좌표 계산
    moments = cv2.moments(bmask)
    # key-value {}로 저장 (m00: ... 면적(area))
    # m00 : 전체 면적(하얀 픽셀 객수)
    m00 = moments['m00']

    # 중심좌표(centroid) 초기화
    centroid_x, centroid_y = None, None

    # 만약 면적이 0이 아니라면 (면적이 존재한다면)
    if m00 !=0:
       centroid_x = int(moments['m10']/m00) # 한 영역의 중심 x 좌표
       centroid_y = int(moments['m01']/m00) # 한 영역의 중심 y 좌표

       # m10, m01 의미 (1차 모멘트)
       # 1차 모멘트 : 이미지 픽셀 같이 이산 데이터(1, 0 등)에서 각 요소의 값(강도) 위치 곱한 것들의 합
       # 이미지 처리에서 1차 모멘트 계산시 모든 픽셀의 위치(x,y)는 (0,0) (맨 왼쪽 상단) 지점부터 거리
       # m10 : x축 방향에 대한 pixel 강도의 합
       # m01 : y축 방향에 대한 pixel 강도의 합
       # 즉, 해당 좌표의 픽셀의 밝기 의미
       # 참고) 0차 모멘트(m00) 어떤 존재(객체) 있냐, 그것의 크기
       # >> 객체의 질량(면적) >> 모든 픽셀 값을 다 더함

       # 중심좌표(centroid) 공식
       # cx(객체의 가로 중심 좌표) = m10 / m00(전체면적)
       # cy(객체의 세로 중심 좌표) = m01 / m00(전체면적)



    # 기본값: 중심 없음
    # >> 초록색을 못 찾으면
    # 왜 -1 을 썼나요? 유효하지 않은 좌표(x좌표, y좌표 >=0)
    # >> 즉, 이미지 화면에 보이는 것 (0,0)~(w,h)
    # center point 없어 >> 객체 못 찾았어
    # >> 객체가 있다면 (중심점 있다면) >> (cx, cy)

    # -1은 이미지 평면에 존재 안하는 숫자([0,255])
    # 좌상단 (0,0)
    crt = (-1, -1)

    # 중심이 계산된 경우
    if centroid_x != None and centroid_y != None:
      crt = (centroid_x, centroid_y)
      # 이미지 위에 중심점 표시(검은색 점)
      cv2.circle(image, crt, 4, (0,0,0), -1)
      # 4: radius(반지름), -1 : 원의 내부를 지정된 색으로 채워라

    cv2_imshow(image) #colab에서 cv2.imshow() 대신 사용

    return crt

# 메인 실행 부분 (동영상 캡처)

if __name__ == '__main__':
    # 프로그램 시작점(이 파일을 직접 실행할 때만 동작)
    capture = cv2.VideoCapture(video_path)

    if not capture.isOpened():
        print('동영상을 열 수 없습니다.')
    else:
        frame_idx = 0

        while True:
            okay, image = capture.read()

            # 비디오가 마지막이면
            if not okay:
                print('영상 끝까지 재생 완료')
                break

            # 초록색 공 추적
            ctr = track(image)
            print(f'Frame {frame_idx}: centroid = {ctr}')

            frame_idx += 1

        capture.release()
```
